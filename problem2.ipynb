```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

mean1 = (-3, 0)
cov1 = np.array([[1.0, 0.8], [0.8, 1.0]])
size = 500
dummy_data1 = np.random.multivariate_normal(mean1, cov1, size=size)
```

```python
x1 = dummy_data1[:, 0]
x2 = dummy_data1[:, 1]

plt.title('Scatter Plot of Dummy Data')
plt.xlabel('X1')
plt.ylabel('X2')
plt.scatter(x1, x2, alpha=0.5)
```
