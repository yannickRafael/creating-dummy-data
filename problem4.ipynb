```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
size = 500
```

```python
mean1 = (-3, 0)
cov1 = np.array([[1.0, 0.8], [0.8, 1.0]])
dummy_data1 = np.random.multivariate_normal(mean1, cov1, size=size)
```

```python
mean2 = (0, -3)
cov2 = np.array([[1.0, 0.8], [0.8, 1.0]])
dummy_data2 = np.random.multivariate_normal(mean2, cov2, size=size)
```

```python
x1 = dummy_data1[:, 0]
x2 = dummy_data1[:, 1]
y1 = dummy_data2[:, 0]
y2 = dummy_data2[:, 1]
```

```python
plt.title('Scatter Plot of Dummy Data')
plt.xlabel('X1')
plt.ylabel('X2')
plt.scatter(x1, x2)
plt.scatter(y1, y2)
plt.legend(["0" , "1"])
plt.show()
```
