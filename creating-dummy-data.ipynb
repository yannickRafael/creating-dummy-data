### Creating dummy data

#### Problem 1: Creating random numbers

```python
import numpy as np

np.random.seed(0)

mean1 = (-3, 0)
cov1 = np.array([[1.0, 0.8], [0.8, 1.0]])
size = 500
dummy_data1 = np.random.multivariate_normal(mean1, cov1, size=size)

# print(dummy_data1)
```

#### Problem 2: Visualization by scatter plot

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Since our dummy_data1 have a (500, 2), we can use the first 
# column for X values and the second for Y values of our scatter plot.
x1 = dummy_data1[:, 0]
x2 = dummy_data1[:, 1]

plt.title('Scatter Plot of Dummy Data')
plt.xlabel('X1')
plt.ylabel('X2')
plt.scatter(x1, x2, alpha=0.5)
```

#### Problem 3: Visualization by histogram

```python
plt.xlabel("X1")
plt.ylabel("Frequency")
plt.title("Histogram of X1")
plt.xlim(-10, 10)
plt.hist(x=x1)
plt.show()
```

```python
plt.xlabel("X2")
plt.ylabel("Frequency")
plt.title("Histogram of X2")
plt.xlim(-10, 10)
plt.hist(x=x2)
plt.show()
```

#### Problem 4: Addition of data

```python
# Creating dummy data 2
mean2 = (0, -3)
cov2 = np.array([[1.0, 0.8], [0.8, 1.0]])
dummy_data2 = np.random.multivariate_normal(mean2, cov2, size=size)

# Plotting both dummy data 1 and 2 in the same scatter plot
x1 = dummy_data1[:, 0]
x2 = dummy_data1[:, 1]
y1 = dummy_data2[:, 0]
y2 = dummy_data2[:, 1]

plt.title('Scatter Plot of Dummy Data')
plt.xlabel('X1')
plt.ylabel('X2')
plt.scatter(x1, x2)
plt.scatter(y1, y2)
plt.legend(["0" , "1"])
plt.show()

```

#### Problem 5: Data combination

```python
combined_data: np.ndarray = np.concatenate([dummy_data1, dummy_data2])
print(f"Dummy data 1 shape: {dummy_data1.shape}")
print(f"Dummy data 2 shape: {dummy_data2.shape}")
print(f"Combined data shape: {combined_data.shape}")
```

#### Problem 6: Labeling

```python
# Creating the labels by repeating the 0's and 1's 500 times each and 
# then concatenating the result to get a 1000 elements array of labels 
labels = np.array([0] * size + [1] * size)
print(f"Labels shape: {labels.shape}")

# Append the labels to the combined data
combined_data_with_labels = np.hstack([combined_data, labels.reshape(-1, 1)])
print(f"Combined data with labels shape: {combined_data_with_labels.shape}")
print("First and last 3 rows:")
print(combined_data_with_labels[0:3])
print(combined_data_with_labels[-4:-1])
```
